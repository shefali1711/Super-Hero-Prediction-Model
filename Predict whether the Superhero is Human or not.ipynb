```python
#checking if there is any null value or missing value in the table
print(df_heroes.isnull().sum())
```

```python
#to know what kind of a column is publisher 
df_heroes['Publisher'].describe()
```

```python
#replacing all the missing or null values by mode
df_heroes['Publisher'].fillna(df_heroes['Publisher'].mode()[0], inplace=True)
df_heroes['Publisher'].isnull().sum()
```

```python

df_heroes['Weight'].describe()
```

```python
#replacing all the missing or null values by mean
df_heroes['Weight'].fillna(df_heroes['Weight'].mean(), inplace=True)
df_heroes['Weight'].isnull().sum()
```

```python
print(df_heroes.isnull().sum())
```

```python
print(df_super_powers.isnull().sum())
```

```python
#converting all the obect type data into numerical data
dummy_hero_col=['Gender','Publisher','Alignment','Eye color','Hair color','Skin color']
df_heroes_dummies = pd.get_dummies(df_heroes, columns=dummy_hero_col)
df_heroes_dummies.head()
```

```python
#converting all the obect type data into numerical data
dummy_power_col=df_super_powers.columns.drop("hero_names")
df_power_dummies = pd.get_dummies(df_super_powers, columns=dummy_power_col)
df_power_dummies.head()
```

```python
#inserting a new column fro human prediction
df_heroes_dummies.insert(106,'Human',0)
```

```python
#catoegrising the hero as human or not
for i in range (0,len(df_heroes_dummies['Race']),1):
    if (df_heroes_dummies['Race'][i]=='Human'):
       df_heroes_dummies['Human'][i]=1
    else:
       df_heroes_dummies['Human'][i]=0
```

```python
df_heroes_dummies.head()
```

```python
#merge both the tables for better predictions
hero_super_powers = pd.merge(df_heroes_dummies, df_power_dummies, left_on=['name'], right_on=['hero_names'], how='inner')
hero_super_powers.head()
```

```python
#dependent and independent variable 
drop_column=['Unnamed: 0','name','Race','Human','hero_names']
x =hero_super_powers.drop(drop_column, axis=1)
y=hero_super_powers['Human']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=400)
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc_pred=rfc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test,rfc_pred))
print('\n')
print(classification_report(y_test,rfc_pred))
```
